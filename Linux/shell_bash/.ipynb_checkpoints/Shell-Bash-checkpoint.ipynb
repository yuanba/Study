# Shell
## 1.什么是Shell
* shell又称计算机壳，是我们的操作系统中为了访问我们的系统内核和我们的相关的应用程序的一种工具，与其说是一种过句，不如说shell其实也是哟中程序设计语言
* shell是一种命令语言也是程序设计语言，分为有两种工作模式，一种是交互式(Interactive)一种是批处理(Batch)模式
* 在众多的shell解释器中，我们这里学习linux下默认的解释器bash有点有以下几种
    * doskey:快速浏览命令历史
    * Tab:自动补全
    * help:完整的联机帮助文档
* Shell & Bash优点
    * 简单，作为知己和系统内核交互的一种命令语言
    * 可移植性强，因为有POSIX标准作为基础，一个shell程序可以在不同的机器上运行
    * 开发简单
    
## 2.基础语法
### 琐碎语法点
* 保存文件的后缀是.sh,在批处理模式下，我们在文件的开头加上如下标记表示我们调用bash解释器解释我们的shell程序
```
#!/bin/bash
```

* 我们使用分号';'来分割我们的命令，可以在一行中用此方法键入多组连续的命令
* 运行shell文件有两种方式
    * 加上可执行权限 - 键入./文件名调用执行(因为我们文件首加上了bash标记表示调用bash解释器默认的执行shell程序)
    ```
    chmod +x name.sh
    ./name.sh
    ```
    * 显示调用解释器执行,这时候我们就算在文件头加上了#！标记也没有用，默认使用命令行制定的解释器执行
        * /bin/bash name.sh
        * bash name.sh
* echo 重要参数
    * -e:转义输出，在我们字符串内部的转义字符转义后输出，默认不转义
    * -n:禁止换行输出，默认换行输出
* 条件语句中的格式：[ \$var1 func \$var2 ]空格必须要注意给足
* p=\$[1\*2] / p=\$((1*1000))  #都可以使用
* 千万要小心，我们在for语句中迭代的对象必须要用\$
* exit \$var 表示退出shell进程
* p=(\$字符串) (定义IFS之后可以更方便的分割)，按照空格分割并初始化字符串
* \`command\`这样的输出可以作为我们的程序代码的左值，返回的不是一个字符串，本质上没有返回值，我们可以给"\`command\`"强制转化成字符串
* 我们完全可以使用这样的形式去初始化构建我们的一个变量
```
p=$(command)
e.g. p=$(ls -la)
```
* 本质上，我们的shell将我们的所有的变量都当做我们的字符串来处理，所以我们可以认为我们的shell的变量都是**字符串类型**
        
### 语法
#### 1.变量
* 定义
    * **定义**和**初始化**或者**赋值**的时候我们不加'\$'标记
    * 变量名和值之间没有空格
    * 名称的第一位是字母
    * 名称中间必须没有空格，可以使用_
    * 不可以使用标点符号和bash的关键字（help可以查询bash的关键字）
* 使用/调用：
```
${var}   #养成加{}的习惯
```
* 只读模式：
只读模式表示我们的变量只有在初始化的时候才可以赋值或者从定义成只读的那一刻器就不允许修改  
    * 只读修饰已经有的变量
    * 初始化只读变量
```
num=1
readonly num
num=2  #Error
readonly data=1
data=2 #Error
```
* 删除模式
```
unset var    #删除变量
unset .f function_name    #删除函数
```
* 变量分类
    * 局部变量：存在在shell脚本内作用域的变量
    * 环境变量：在整个操作系统下，所有程序（包括shell在内）的所有的程序都可以访问到的变量（旺旺表示程序的必要的地址的重要信息）
    * shell变量：上面两种
* 特殊变量（特殊标记符）
    * '\$\$'表示当前的进程在操作系统中分配的ID  
    * '\$0'表示当前执行的shell脚本的名称  
    * '\$n'表示第几个命令行参数 n:[1,2,3..n]如上所示，0号表示脚本名称
    * '\$#'表示命令行参数的个数
    * '\$*'所有命令行参数
    * '\$@'所有命令行参数
    * '\$?'上条命令的退出状态和函数的返回值（Int类型）  
    区分 \$* 和 \$@  
    1.共同点：包含的所有参数不含脚本名称的0号参数  
    2.不同：引号输出的情况下
        * '\$*':一个整体输出
        * '\$@':分开输出
        
#### 2.shell替换
* 转义字符
    * \\ - \
    * \a - Bling
    * \r - 回车
    * \n - 换行
* 命令替换
\`\`符号可以将我们的执行的要现实到标准输出的内容一**字符串**的形式返回,bin提供了赋值给变量的可能性
```
p=`[command]`
```
* 变量替换
    * \${var} 变量输出
    * \${var:-word} word是默认值，var不存在则返回word的值，不修改var
    * \${var:=word} word是默认值，var不存在则返回word的中，并修改var
    * \${var:?message} message是消息，var不存在，message动如标准错误流中（注意此时的shell程序运行在次中断）
    * \${var:+word} var存在输出word默认消息，不存在不出输出
   
#### 3.运算
1 bash本身不支持很强的运算，但是我们可以采用辅助性的其他的程序实现我们这方面的缺陷，一个比较有名的就是**expr**程序  
2 expr要点  
* 数据和运算符之间必须要有空格
* 调用
        * `expr ...` 将计算作为一个命令，返回一个结果（字符串，数字的字符串可以就理解是数字）
        * expr ....直接交互式调用
* expr接口：
     * +,-,*,/,%,<=,=,>=,!=,<,>
     唯一需要注意的就是*,<,>,<=,>=需要转义
     * index，length，match，substr，：
         * index STRING PATTERN :检查pattern是否存在于我们的STRING中,存在返回一个非0，否则返回0  
         ```
         expr index 'Something in here' 'in'   # result > 0
         ```
         * length STRING :检查STRING字符串长度
         ```
         expr length 'lantian'   # result = 7
         ```
         * match STRING PATTERN / STRING : PATTERN  正则表达式匹配，匹配到返回一个非负数，否则返回0
         * substr STRING a b 从STRING字符串中，从a开始（1计算起点）,割出长为b的子串
         * expr $var1$var2 - 字符串拼接 
* 数字关系运算：
    * -eq:相等
    * -ne：不等
    * -gt：>
    * -lt:<
    * -ge:>=
    * -le:<=
* 字符串关系运算
    * =：字符串相等
    * !=：字符串不等
    * $var : 单独存在如果串空（长为0）则表达式false，否则true
* 文件测试：
    * -e：是否存在该文件
    * -s:该文件是否为空,不空是true
    * -x:文件是否有执行的权限
    * -w：文件时否有可写权限
    * -r：文件是否有可读权限
    * -f：文件是否是普通的文件
    * -d：文件是否是目录文件
* 逻辑运算：
    * ==:等 - 针对数字
    * !=：不等 - 针对数字
    * !:取反
    * -o：或
    * -a：与

#### 5.shell字符串操作
1 shell中最重要的两种数据类型只有 数字 字符串  
2 字符串的定义
    * 单引号：原样输出，不进行变量替换
    * 双引号或者没有引号：可以替换输出
3 字符串常用操作：
* 拼接：
    * 双引号串内调用变量实现   
    ```
       name=lantian
       greet="hello,\$name"
    ```
    * expr
      ```
       a=1;b=2;
       p=`expr \$a\$b`
      ```
* 字符串长度
    * expr length
    * ```string=lantian;\${#string};```
* 字符串分片
    * \${var:up:length}
    * expr substr a b - 见上，从1开始索引
    
#### 6.shell数组
* 只能定义一维数组，不能定义多维数组,数组从0开始索引，数组长度无限，并且可以给没有元素的位置赋值，并不用考虑前面是不是都有值
* 定义数组：
    array=（var0 var1 var2 .. ）空格分开或者换行分开
    数组的类型和python中的list一样
* 数组分片
   ```
   a=(1 2 3 4 5)
   echo \${a[@]:begin:length}
   ```
* 读取下表：
```
echo ${var[n]}
```

* 获取所有元素：按序将我们的存在的下表进行提取

```
echo ${var[\*]}
echo ${var[@]}
```

2.输出echo/printf  

详细化printf:
printf和我们的C语言中的printf是几乎一样的  
```
printf format-string [arguments .. ]
```
arguments使用空格分开并按顺序，我们的printf记住要加换行哦，和echo不一样，printf默认不换行  
在format中支持的有%c,%d,%s,%f,%e,%E(e,E科学计数法)，C语言使用的语法都成立，并且如果我们的参数不够替换，%s默认是空串，%d默认是0

#####  8.4函数
* 定义：
```
[function] function_name()
{
    command ...
    command ...
    return value   #没有return语句，以最后一条命令的输出最为返回值
}
```
* 返回值
    * 成功返回0,失败返回一个非0值
    * 返回值不可以是非数字，否则调用的时候会报错
* 调用(含参数调用)
```
$fun var1 var2 var3 ...
```
* 函数参数： - 同命令行的参数的语法格式一模一样
    * \$# \$n \$0(结果是/bin/bash) \$* \$@
    
##### 8.5输入输出重定向
* command > file : 覆盖式重定向写入
* command >> file ： 追加式重定向写入
* command < file ： 输入重定向
* 0,1,2分别代表我们的输入文件描述，输出文件描述，错误文件描述符
* command > file 2>&1 : 错误重定向和处处重定向合并
* <> : 同时使用，屏蔽掉键盘输入和屏幕输出
* > /dev/null : 禁止输出
*表示将下面的块作为输入重定向给命令去执行
```
command << tag
command1...
command2...
command3...
tag
```
