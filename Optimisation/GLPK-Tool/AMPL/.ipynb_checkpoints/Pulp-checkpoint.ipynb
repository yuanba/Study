对于我们的一直的线性规划的问题，我们一直了一种非常优秀的求解器叫做GLPK，Python内部对这些一类的非常优秀的求解器提供了一个叫做pulp的模块作为接口，并且十分方便的为我们的**数据的导入**和**计算**以及**约束的构成**我们的提供了非常便利的途径
这种便利体现在一下几个方面上
* Python拥有非常强大的第三方库，可以轻松的实现多种文件的数据读取,甚至在为了验证模型的正确性上，我们甚至可以快速的生成随机数据
* Python的表达式清晰简单，可以快速的构建我们需要的参数
* 屏蔽了AMPL的语法细节

相对于AMPL   
1.集合我们采用列表的形式存储     
2.参数可以采用字典的形式来存储，其中字典的键是被索引的集合的名称，值是数据提供给参数的值，我们的一个字典可以理解成是一个参数矩阵的一列，我们一个字典就是一列的参数的表示,**当然我们的参数依旧可以用Python的列表等其他任何合适的数据类型都可以**   
3.目标函数     
4.约束表达式      
5.变量我们可以使用LpVariable.dicts()方法将列表转化成字典，这时候我们和参数统一了   


表达式的书写中，我们的主要的核心部分就是使用字典的调用语句，并且我们的约束的现行表达式中可以使用python变量   
我们在表达式中可以尽可能的使用一些python的运算的符号和函数  
**记住我们的就算在复杂的表达式也可以用列表推倒式/元组推倒式代替**

基本语法
* 定义问题
```
p = LpProbelm('Question_name',LpMaximize)   #Or LpMinimize
```
* 定义变量
```
x = LpVariable('变量描述名',lowwer,upper,类型) #类型是AMPL中的定义的，lowwer和upper不指定则认为在-inf～inf之间
```
* 增加目标函数
```
p += 变量表达式，‘目标函数描述’
```
* 增加约束
注意约束中的符号只有==(=),>=,<=三种
```
p += 变量表达式（约束），‘约束描述’
```
* 问题写入指定的.lp文件
```
p.writeLP('指定文件名.lp')
```
* 问题解决
```
p.solve()
```
* 问题状态查询
```
#所有的问题求解状态都在LpStatus字典中，我们只需要用
print(LpStatus[p.status])
```
* 问题变量查询
```
p.variables()返回一个所有变量组成的列表，我们可以迭代打印
```
* 问题求解结果
```
value(p.objective)
```
* LpVariable
        * LpVariable(name,lowbound,upbound,变量的类型cat)
        * dicts('前缀'，列表，变量的默认值，变量的类型) - 返回一个有\'前缀_\'为前缀，之后的变量名为后缀的一个新的值（变量），key是我们的原列表中的字符串，目的是为了在表达式中的在集合的索引上和我们的参数的字典统一（变量使我们的单独定义额对象这里就不用考虑）
        * matrix('前缀'，（列表组）,每一个变量的上界，每一个变量的下界，参数类型)
        上面的三种类型其实非常好理解，第一种就是调用构造函数简单的生成单个的变量  
        第二种是利外界定义的字符串序列（list等）作为键值生成变量字典在使用的时候利用索引我们的集合（外界的序列）实现调用变量   
        但是后者实际上是生成一种变量的矩阵，列表组实际上是一个元组，元组内的元素是序列的可迭代对象，后面的声明是针对矩阵中的每一个的变量的并且元组只有一个可迭代对象代表是一个1×n的矩阵，两个就是n×m的矩阵，一次类推

* lpSum() - 內建函数，对我们的列表求和病生成一个标准的pulp表达式  
其实我们都可以使用python内置的循环语句和表达式来实现这个类似的所有的表达式的功能：列表推倒是，math函数，for语句多次迭代产生同类型的约束语句等等
