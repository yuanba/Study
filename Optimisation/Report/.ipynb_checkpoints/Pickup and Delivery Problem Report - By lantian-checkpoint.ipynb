# 问题描述
1.已知一个一个地图，地图上有n个点，每个点有一个权值，权值可正可负  
2.一位送货员，起始的负重和最终的负重都是0，在途径一个点之后，获取改点的权值作为负重，且负重必须是非负数  
3.送货员每个点都要遍历一次且仅一次，记录每条路径上的负重  
4.求负重总和的最小值  

# 模型建立

# 可行的邻域搜索方案
首先，我们需要先了解一下局部搜索的算法　　  
1. 如果我们采用精确算法求解该问题，(在不考虑是否该序列合法的前提下)我么需要至少考虑$O(n!)$中情况显而易见，在数据量稍微大的时候我们琼剧的思路就会变得异常的低效甚至没有丝毫的用处    
2. 局部搜索，我们稍微的缩短的我们的**解空间**的范围，在部分区域内我们重新更新数据的顺序(注意：不是全局的更新)，在保证快速的同时尽量多的优化我们的初始解    
3. 对邻域搜索的方案中，我们的搜索的策略非常的多，如下，我们罗列一些思路并进行简要的解释(核心的解释我放在总结的图表中进行)  
   在这里我只对前两种的搜索策略进行了实现，所一我给出了伪代码详细表述，之后的方案我仅做分析(模拟退火算法的一样)，请谅解
    * One-pass：  
    该方案，老师上课简要的介绍过，我们选择随意的一段序列区间(该序列区间意义和我们的模型中的是一致),随机打乱我们的顺序   
    如果我们成功的优化了当前的解，我们更新否则继续   
        * 大致时间复杂度:　$O(m!)\ \ \ m:the length of the shuffle space$
        * 伪代码：  
            $def\ onepass:$  
            $Input:$  
            $\ \ \ \ seq\ -\ the\ sequence\ which\ save\ the\ order\ of\ the\ city$  
            $\ \ \ \ n\ -\ the\ number\ of\ the\ city$  
            $\ \ \ \ w\ -\ the\ weight\ of\ each\ city$   
            $Output:$   
            $\ \ \ \ best\ -\ the\ best\ solution\ of\ this\ turn\ of\ onepass\ function$  
            $Side\ effect:$  
            $\ \ \ \ the\ sequence\ maybe\ changed\ after\ this\ function$  
            $\ \ \ \ \ \ \ \ \ \ \ $  
            $best\ :=\ count(seq,w)$  
            $while\ Limite:$  
            $\ \ \ \ begin\ :=\ Random(1,n)$
            $\ \ \ \ end\ :=\ Random(begin,n)$    
            $\ \ \ \ shuffle(begin,end)$    
            $\ \ \ \ if\ check(seq)\ ==\ True:$  
            $\ \ \ \ \ \ \ \ best\ :=\ count(seq,w)$  
            $\ \ \ \ else:$  
            $\ \ \ \ \ \ \ \ Reshuffle(begin,end)$  
            $\ \ \ \ Limite decrease$  
            $Return\ best$
    * Two-choose：  
    该方案，我们随机的选取两个城市，然后交换位置，判断能否对现在的最优值进行优化  
    相对与上面的方案，该方案的优点就是搜索的空间小，所以优化的速度非常的块，经过实际的检验效果来看，该方案优化效果和第一种相比大致相同，但是我们的优化的速度非常的块  
        * 时间复杂度：$O(n^2)$  
        * 伪代码:  
            $def\ twochoose:$  
            $Input:$  
            $\ \ \ \ seq\ -\ the\ sequence\ which\ save\ the\ order\ of\ the\ city$  
            $\ \ \ \ n\ -\ the\ number\ of\ the\ city$  
            $\ \ \ \ w\ -\ the\ weight\ of\ each\ city$   
            $Output:$   
            $\ \ \ \ best\ -\ the\ best\ solution\ of\ this\ turn\ of\ onepass\ function$  
            $Side\ effect:$  
            $\ \ \ \ the\ sequence\ maybe\ changed\ after\ this\ function$  
            $\ \ \ \ \ \ \ \ \ \ \ $  
            $best\ :=\ count(seq,w)$  
            $while\ Limite:$  
            $\ \ \ \ begin,end\ :=\ Sample(seq,2)$   
            $\ \ \ \ swap(begin,end)$    
            $\ \ \ \ if\ check(seq)\ ==\ True:$  
            $\ \ \ \ \ \ \ \ best\ :=\ count(seq,w)$  
            $\ \ \ \ else:$  
            $\ \ \ \ \ \ \ \ Reshuffle(begin,end)$  
            $\ \ \ \ Limite decrease$  
            $Return\ best$
    * EBSR:
    老师上课的的时候，讲述过EBSR的优化的思路
        * 将序列截取一部分
        * 滑动截取序列的一部分不断的查找可优化的最优值
        * 返还初始解
        * 缩小截取的上界，重复上述的操作　
    * EFSR:
    同上的思路，只不过我们采用相反的方向，可能在默写数据的情况下，该思路的优化效果要更加的优秀

# 总结
## 分析贪心邻域搜索：
不论是**Onepass**,**Twochoose**,**EBSR**,**EFSR**这里面的哪一种方法我们都需要知道，我们的在整个搜索的过程中，都是在一个方法决定的固有的邻域内去搜索我们的最优值，客观的来说，这样的搜索方式会导致一种特殊的情况，我们每次都是选取了当前的最优的情况，然后以这个最优的解出发去找下一个最优解  
这好比是一颗决策树，我们每一次都需要站在决策树的某一个节点上，并且这些搜索策略决定了我们只能只能站在当前的所有后继节点中最优的哪一个上，很容易陷入某一个具体的分支中而永远无法找到全局最优解，有可能全局最优解存在于我们另一个分支上，当我们从决定下一个后继的那一刻起就注定无法达到那个更有的值，换句话来说，我们陷入了一个局部最优解  
**有可能这个局部最优解就是全局最优，但有可能会是一个非常差的情况**，相比来说，我们找到一个更好的解的情况在这些贪心的算法中并不是很高效，所以我们需要引入另一种更优秀的算法  

## 模拟退火算法改进：
1. 不同于上述的贪心的算法，我们的模拟退火算法采用另一种思想去优化上面的算法
    * 有时候，我们短暂的后退是为了更好的前进，我们的模拟退火算法允许我们以一定的概率接受一个并不是最优的解，甚至是一个更坏的解，去实现我们有选择行的随机的回溯到我们的之前的分支上，去尝试一些可能更接近全局最优解的优良解
    * 模拟工业的退火的原理，我们在随着温度的不断的下降的过程中，将我们这种接收更坏的解的概率逐渐的缩小（为了之后的规约到最优解上），实验证明，这种情况下，我们的搜索最优解的效率会非常的高效
2. 算法描述和伪代码:  
    $def\ SA:$    
    $Input:$  
    $\ \ \ \ Schedule\ -\ the\ mapping\ of\ producting\ the\ decreasing\ T\ function$  
    $\ \ \ \ n\ -\ the\ number\ of\ the\ city$  
    $\ \ \ \ w\ -\ the\ weight\ of\ each\ city$
    $\ \ \ \ seq\ -\ the\ sequence\ of\ the\ city$
    $\ \ \ \ now\ -\ the\ init\ solution\ of\ the\ question$
    $\ \ \ \ next\ -\ the\ next\ solution\ from\ the\ now\ point$  
    $Output:$   
    $\ \ \ \ best\ -\ the\ best\ solution\ of\ this\ turn\ of\ SA\ Algorithm$  
    $Side\ effect:$  
    $\ \ \ \ the\ sequence\ maybe\ changed\ after\ this\ function$  
    $\ \ \ \ \ \ \ \ \ \ \ $  
    $now\ :=\ create\_init()$    
    $for\ t\ :=\ 1\ to\ \infty$    
    $\ \ \ \ T\ :=\ Schedule(t)$  
    $\ \ \ \ if\ T\ =\ 0:Return\ now$    
    $\ \ \ \ next\ :=\ create\_from(now)$       
    $\ \ \ \ \Delta E\ :=\ next\ -\ now$  
    $\ \ \ \ if\ \Delta E\ is\ better:$  
    $\ \ \ \ \ \ \ \ now\ =\ next$  
    $\ \ \ \ else:$    
    $\ \ \ \ \ \ \ \ now\ =\ next\ with\ the\ possiblity\ of\ e^{\Delta E\ /\ T}$      
    $Return\ best$

```
# The Mod Solve the problem which is more like the TSP problem
# But which has more constrains

/* Set */
#set place 'Place which the milkman should go \$every\$ \$only once\$';

/* Param */
param n;   # The number of the place
param w {i in 1..n};   # The cost of the place

/* var */
var l {i in 0..n} >=0;# 'l[i] means after the place how many milk milkman last' The most important variable
var x {i in 1..n,j in 1..n} binary;

/* Objective function */
minimize ans: sum{i in 0..n} l[i];

/* Constraints */
s.t. x_row{i in 1..n} : sum{j in 1..n} x[i,j] = 1;
s.t. x_col{j in 1..n} : sum{i in 1..n} x[i,j] = 1;
s.t. l_fisrt : l[0] = 0;
s.t. l_end : l[n] = 0;
s.t. l_cons {i in 1..(n-1)} : l[i] = l[i-1] + sum{j in 1..n} x[j,i]*w[j];
s.t. l_neg {i in 2..(n-1)} : l[i] >= 0;

end;
```

### 1.2 自我实现的两种算法的思路和具体代码：
#### 1.2.1 事后优化策略 - 引入DFA（有限状态自动机）
在贴代码之前，我需要首先讲解一下我事后优化的思路：
我们会发现，实际上，在优化的过程中，我们是随机的优化，但是实际上，我们完全可以对一些特有的情况进行具体的分析从而进一步优化我们的结果  
* 序列中两个相邻的城市的权值都是正的时候：
    这种时候我们会发现，我们如果如果将小权值的城市先遍历，后遍历大权值的城市是可以将我们的总权值降低的 
* 序列中两个相邻的城市的权值都是负数的时候：
    这时候我们会发现，如果在满足条件的情况下，我们先遍历最小的城市，后遍历权值大额城市也是可以将我们的总权值降低的
* 序列中一正一负相邻的时候：
    这种时候，我们先遍历负权值的城市明显会优化我们的总权值
基于以上的几种优化策略，很明显符合正则表达式的要求，可以使用有限状态自动机(DFA)来实现扫描和进一步优化  
注意如果出现了连续的以上的优化的情况，我选择的是递进的方式，这样的优点是可以加快我们的事后优化的速度

#### 1.2.2 Onepass优化：
```
def solve(seq , data , step = 5 , w = 5):
    '''
    This function try to solve the init sequence and create the local Optimisation result.
    If I use the SA Algorithm to upgrade the Climbing-Hill Algorithm , I can enlarge the step and decrease the w.
    Input:
        seq - the init sequence of the city
        data - the weight of each city
        step - the step of the onepass moving
        w - the length of the interval
    Output:
        (ptime , best , origin , seq) - the local Optimisation of the result
    Side affect:
        seq - seq maybe changed after every search , but we will return the most Optimisation result which is satified
        In the Python , It must not be changed , becaouse i use the  = operator.But it is also risky.
    '''
    a = time()
    n = len(data)  # Calculate the size of the city.
    origin = best = count(seq , data)
    limite = 1
    for i in range(1,w+1):
        limite = limite * i
    for i in range(1,n+1,step):
        begin = i
        end = i + w - 1
        if end > n:break
        pause = seq[begin:end + 1]
        now = deepcopy(seq)
        for k in range(limite):
            shuffle(pause)
            now[begin:end + 1] = pause
            if check(now , data) == False:
                new = count(now , data)
                if new < best:
                    best = new
                    seq = now
                    break
    b = time()
    return (b - a,best,origin,seq)

```
def solve(seq , data , maxp):
    '''
    This function is anthor method which can compared with the method onepass . In this case , I choose two random sequnece's number and try to swap them . And if the new point is better than now , I try to choose it as the now point.
    But , We should remember that because I use the builtin.method from random.sample (Which can return n(You decide) random number from the sequence) so I really need to choose the max number of trying to sqap two random number . So I set a input function to hold this tag.
    If I try to use the SA Algorithm to upgrade the Climbing-Hills Algorithm , I can decrease the maxp.
    Input:
        seq - The sequence I got from init.
        data - The weight of each city.
        maxp - The number of the max to hold decide the times of our iterations.
    Output:
        (ptime , best , origin , seq) - the local Optimisation of the result
    Side affect:
        seq - sequence maybe changed after the fucntion . But  We only care about the optimal. So it doesn't matter.
    '''
    a = time()
    origin = best = count(seq , data)
    #maxp = int(input('Please input the maximize numher: '))
    for i in range(maxp):
        lucky = sample(seq , 2)
        seq[lucky[0]] , seq[lucky[1]] = seq[lucky[1]] , seq[lucky[0]]
        if check(seq,  data) == False:
            new = count(seq , data)
            if new < best:
                best = new
                continue
        seq[lucky[0]] , seq[lucky[1]] = seq[lucky[1]] , seq[lucky[0]]
    b = time()
    return (b - a , best , origin , seq)

def fun_onepass(seq , data):
    n = len(data)
    # Define the shuffle space (begin , end)
    begin = randint(1, n - 2)
    end = randint(begin + 1 ,n - 1)
    best = count(seq , data)
    limite = 1
    for i in range(begin , end + 1):
        limite = limite * i
    for i in range(limite):
        pause = seq[begin:end + 1]
        tty = deepcopy(pause)
        shuffle(pause)
        seq[begin:end + 1] = pause
        if check(seq , data) == False:
            return count(seq , data)
        else:seq[begin:end + 1] = tty
    return 0

def fun_twochoose(seq , data):
    n = len(data)
    # Define the change position
    limite = n*(n-1)//2
    for i in range(limite):
        begin , end = sample(range(1,len(seq)) , 2)
        seq[begin] , seq[end] = seq[end] , seq[begin]
        if check(seq , data) == False:
            return count(seq , data)
        else:seq[begin] , seq[end] = seq[end] , seq[begin]
    return 0
